```python
#importing neccesary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset from CSV
data = pd.read_csv(r"C:\Users\Mounika\OneDrive\Desktop\population_dataset_50.csv") 
print(data)
```

```python
# Histogram for Age (continuous variable)
if 'Age' in data.columns:
    plt.figure(figsize=(10, 6))#figure height and width
    plt.hist(data['Age'], bins=20, color='orange', edgecolor='black')#generating a histogram for column age with 20bins of orange colour and black border
    plt.title('Distribution of Age in Population using "HISTOGRAM"')#adding a title to histogram
    plt.xlabel('Age')#age on x-axis
    plt.ylabel('Number of People')#no.of people on y-axis
    plt.grid(axis='y', alpha=0.75)#to add grid along y-axis(0.75 is slightly visible lines)
    plt.show()#plot the graph
else:
    print("Column 'Age' not found in dataset.")
```

```python
# Bar Chart for Gender (categorical variable)
if 'Gender' in data.columns:
    plt.figure(figsize=(8, 5))
    gender_counts = data['Gender'].value_counts()#to count the no.of people of each gender
    print(gender_counts)
    gender_counts.plot(kind='bar', color=['lightpink','lightgreen'], edgecolor='black')
    plt.title('Gender Distribution in Population using "BAR CHART"')
    plt.xlabel('Gender')
    plt.ylabel('Number of People')
    plt.grid(axis='y',alpha=0.25)
    plt.xticks(rotation=0)#to place lables on x-axis in a proper way without any rotation
    plt.show()
else:
    print("Column 'Gender' not found in dataset.")
```

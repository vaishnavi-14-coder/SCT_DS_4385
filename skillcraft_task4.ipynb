```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import warnings
warnings.filterwarnings("ignore")

# Load the data
file_path =r"C:\Users\Mounika\OneDrive\Desktop\skillcraft\Road_Accidents_2017-Annuxure_Tables_3.csv"
df = pd.read_csv(file_path)
# Clean column names
df.columns = df.columns.str.strip()
#choose relevant year column
year = 2017
column_name = f'State/UT-wise Total Number of Persons Killed in Road Accidents during - {year}'
#Drop rows where the value is missing 
df = df[['States/UTs',column_name]].dropna()
# Sort by number of deaths
df_sorted = df.sort_values(by=column_name, ascending=False)

# Plot the time series
plt.figure(figsize=(15, 8))
sns.lineplot(x=column_name,y='States/UTs',data=df_sorted)
plt.title('Road Accidents Deaths by States/UTs in {year}')
plt.xlabel('Deaths')
plt.ylabel('States/UTs')
plt.tight_layout()
plt.show()
```
